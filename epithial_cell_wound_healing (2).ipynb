```python
# ------------------------------------------------------------
# 8. Create professional 3D animation with labeled axes
# ------------------------------------------------------------

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

def update(frame_idx):
    ax.clear()

    u_f, I_f, B_f, S_f = frames[frame_idx]
    Z, colors = render_frame(u_f, I_f, B_f, S_f)
    phase = compute_phase(u_f, I_f, S_f)

    # Calculate wound closure percentage
    current_wound_area = np.sum(u_f < 0.5)
    initial_wound_area = np.sum(W0 == 1)
    wound_remaining = 100 * current_wound_area / initial_wound_area
    wound_closed = 100 - wound_remaining

    # 3D surface
    ax.plot_surface(
        X, Y, Z,
        facecolors=colors,
        rstride=1,
        cstride=1,
        linewidth=0,
        antialiased=True,
        shade=False
    )

    # Professional title
    ax.set_title(
        "3D Multiphase Skin Wound Healing Simulation\n"
        f"Frame {frame_idx} | {phase} | Wound closure: {wound_closed:.1f}%",
        fontsize=13,
        pad=18
    )

    # Axis limits
    ax.set_xlim(-1, 1)
    ax.set_ylim(-1, 1)
    ax.set_zlim(-0.60, 0.15)

    # Axis labels
    ax.set_xlabel("x-position across tissue surface\n(normalized units)", fontsize=10, labelpad=10)
    ax.set_ylabel("y-position across tissue surface\n(normalized units)", fontsize=10, labelpad=10)
    ax.set_zlabel("relative tissue surface height", fontsize=10, labelpad=10)

    # Tick marks
    ax.set_xticks([-1, -0.5, 0, 0.5, 1])
    ax.set_yticks([-1, -0.5, 0, 0.5, 1])
    ax.set_zticks([-0.5, -0.25, 0])

    ax.tick_params(axis='both', which='major', labelsize=8)
    ax.tick_params(axis='z', which='major', labelsize=8)

    # Better 3D proportions
    ax.set_box_aspect((1, 1, 0.35))

    # Camera angle
    ax.view_init(elev=42, azim=45)

    # Light grid for professional look
    ax.grid(True)

    # Add explanatory annotation inside the plot
    ax.text2D(
        0.02, 0.94,
        "Model variables:\n"
        "u = epithelial cell density\n"
        "I = inflammation signal\n"
        "B = vascular support\n"
        "S = scar remodeling",
        transform=ax.transAxes,
        fontsize=9,
        verticalalignment="top",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.75)
    )

    ax.text2D(
        0.70, 0.06,
        "Color interpretation:\n"
        "red = open/inflamed wound\n"
        "dark red branches = angiogenesis\n"
        "pale region = scar tissue\n"
        "skin tone = healed tissue",
        transform=ax.transAxes,
        fontsize=9,
        verticalalignment="bottom",
        bbox=dict(boxstyle="round", facecolor="white", alpha=0.75)
    )

    return []

anim = FuncAnimation(
    fig,
    update,
    frames=len(frames),
    interval=110,
    blit=False
)

plt.close(fig)

HTML(anim.to_html5_video())
```
